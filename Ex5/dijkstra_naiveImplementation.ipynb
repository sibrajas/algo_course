```python
    import copy
    class Graph:
        def __init__(self):
            self.vertices={}
        def add_vertex(self,vertex1):
            self.vertices[vertex1]=[]
        def add_edge(self,vertex1,vertex2,edgelen):
            if vertex1 in self.vertices.keys():
                self.vertices[vertex1].append([vertex2,edgelen])
            else:
                self.vertices[vertex1]=[vertex2,edgelen]
        def disp(self):
            for vertex in self.vertices.keys():
                print "-------------Edges of %d------------" %vertex
                for edge in self.vertices[vertex]:
                    print edge
        def BFS_reachable(self,source):
            explored=[]
            queue=[]
            queue.append(source)
            while len(queue)>0:
                vertex=queue[0]
                explored.append(vertex)
                #print "Status of the queue with the vertex processed
                #print queue,vertex
                queue.pop(0)
                if vertex not in self.vertices.keys():
                    continue
                for vertex2,edgelength in self.vertices[vertex]:
                    if vertex2 not in explored:
                        if vertex2 not in queue:
                            queue.append(vertex2)
            #print "Reachable nodes from %d" % source
            #print explored
            return explored

        def dijkstras(self,source):
            selfvertices=self.vertices.copy()
            X=[];A={};path={}
            #V=[v for v in selfvertices.keys()]
            V=self.BFS_reachable(source)
            A[source]=0
            path[source]=[source]
            X.append(source)
            V.remove(source)
            while len(V)>0:  #O(mn)
                djgreedyscore=100000
                for vertex in X:
                    for vertex2,edgelength in selfvertices[vertex]:
                        if vertex2 in V:
                            if djgreedyscore>A[vertex]+edgelength:
                                djgreedyscore=A[vertex]+edgelength
                                minedge=[vertex,vertex2,edgelength]
                X.append(minedge[1])
                path[minedge[1]]=copy.copy(path[minedge[0]])
                path[minedge[1]].append(minedge[1])
                V.remove(minedge[1])
                A[minedge[1]]=djgreedyscore
                selfvertices[minedge[0]].remove([minedge[1],minedge[2]])
            #print "Dijkstras paths from source vertex %d" %source
            #print path
            return A,path
    if __name__=='__main__':
        g=Graph()
        g.add_vertex(1)
        g.add_vertex(2)
        g.add_vertex(3)
        g.add_vertex(4)
        g.add_edge(1,2,1)
        g.add_edge(1,3,4)
        g.add_edge(2,3,2)
        g.add_edge(2,4,6)
        g.add_edge(3,4,3)
        #g.BFS_reachable(4)
        #g.disp()
        x,p=g.dijkstras(3)
        g1=Graph()
        fd=open("dijkstraData.txt")
        lines=fd.readlines()
        for line in lines:
            line=line.strip().split()
            vertex1=int(line[0])
            g1.add_vertex(vertex1)
            for edge in line[1:]:
                edge=edge.split(',')
                [vertex2,edgelen]=map(int,edge)
                g1.add_edge(vertex1,vertex2,edgelen)
        #g1.disp()
        s,p=g1.dijkstras(1)
        for i in [7,37,59,82,99,115,133,165,188,197]:
            if i not in s.keys():
                s[i]=1000000
            print 'Destination vertex is %d' %i
            print str(s[i])
            print p[i]
```
