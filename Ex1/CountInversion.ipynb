```python
%matplotlib inline
import seaborn as sns; sns.set()
import matplotlib.pyplot as plt
import numpy as np

n=np.linspace(-1,1000,1000)
plt.axis((0,100,0,50))
colors=['b','g', 'r', 'c', 'm', 'y', 'k']
fns=['np.sqrt(n)','np.power(10,n)','np.power(n,1.5)','np.power(2,np.power(np.log(n),0.5))','np.power(n,(5.0/3))']
plt.scatter(n,np.log(np.sqrt(n)),c=colors[0],label="a "+fns[0],marker='o')
plt.scatter(n,np.log(np.power(10,n)),c=colors[1],label="b "+fns[1])
plt.scatter(n,np.log(np.power(n,1.5)),c=colors[2],label="c "+fns[2])
plt.scatter(n,np.log(np.power(2,np.sqrt(np.log(n)))),c=colors[3],label="d "+fns[3])
plt.scatter(n,np.log(np.power(n,(5.0/3))),c=colors[4],label="e "+fns[4])
plt.legend(loc="best")
```
